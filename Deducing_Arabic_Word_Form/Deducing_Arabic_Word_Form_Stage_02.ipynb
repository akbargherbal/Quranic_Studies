```python
def form_i(word):
    combs = list(combinations(list(range(len(word))), 3))

    combs = [list(zip(i, 'فعل')) for i in combs]

    list_container = []
    for com in combs:
        text = list(word)
        text[com[0][0]] = com[0][1]
        text[com[1][0]] = com[1][1]
        text[com[2][0]] = com[2][1]
        text = ''.join(text)
        list_container.append(text)
    return list_container
```

```python
with open('../DATASETS/Quran_Aya_Per_Line.txt', encoding='utf-8') as f:
    lines = [line.strip() for line in f.readlines()]
    set_words = [[word.strip() for word in line.split()] for line in lines]
    set_words = set(chain(*set_words))
```

```python
df_01 = pd.DataFrame(data=(list(set_words))).rename(columns = {0:'WORD'})
```

```python
df_01['POSS_FORM_I'] = df_01.WORD.apply(lambda x: form_i(x))
```

```python
def try_get_word_form(df_):
    pool = df_['POSS_FORM_I']
    potential = df_['POSS_FORM_FILTER']
    
    container = []
    for pot in potential:
        for item in pool:
            if pot in item:
                container.append(pot)
    return list(set(container))
```

```python
df_01['POSS_FORM_FILTER_I'] = df_01.apply(lambda x: try_get_word_form(x), axis=1)
```
